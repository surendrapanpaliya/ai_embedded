# Supervised Learning Algorithms

    For this example, we'll use a synthetic dataset to perform a regression task with Linear Regression and a classification task with Logistic Regression, Decision Tree, and Random Forest models.

```python
#Step 1: Import Libraries

# Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import root_mean_squared_error, accuracy_score, classification_report, confusion_matrix
import seaborn as sns
```

```python
#help(LinearRegression)
```

## Step 2: Create a Synthetic Dataset


    1. For a regression problem (predicting a continuous target)
    2. For a classification problem (predicting a categorical target)
    1. Regression Dataset
    # Create a synthetic regression dataset

```python
np.random.seed(42)

X_regression = 2 * np.random.rand(100,1)  # 100 samples, 1 feature

print(X_regression[:5])

y_regression = 4 + 3 * X_regression + np.random.rand(100,1)  # Linear relationship y = 4 + 3X + noise

print(y_regression[:5])

# Visualize the regression dataset
plt.scatter(X_regression, y_regression, color='blue', alpha=0.6)

plt.title("Synthetic Regression Dataset")

plt.xlabel("Feature X")

plt.ylabel("Target y")

plt.show()
```

## 2. Classification Dataset

```python
# Create a synthetic classification dataset

X_classification = np.random.rand(100, 2)  # 100 samples, 2 features
y_classification = (X_classification[:, 0] + X_classification[:, 1] > 1).astype(int)  # Binary classification based on sum of features

# Visualize the classification dataset
plt.scatter(X_classification[:, 0], X_classification[:, 1], c=y_classification, cmap='viridis')
plt.title("Synthetic Classification Dataset")
plt.xlabel("Feature X1")
plt.ylabel("Feature X2")
plt.show()
```

## Step 3: Supervised Learning Algorithms Implementation
    
### 1. Linear Regression
    Linear Regression is used to predict continuous values. We’ll train it on the regression dataset.

```python
# Split data into training and test sets
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_regression, y_regression, test_size=0.3, random_state=42)

# Initialize and train the Linear Regression model
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train_reg, y_train_reg)

# Make predictions
y_pred_reg = linear_reg_model.predict(X_test_reg)

# Evaluate the model
mse = root_mean_squared_error(y_test_reg, y_pred_reg)
print(f"Linear Regression Mean Squared Error: {mse}")

# Visualize the regression line
plt.scatter(X_test_reg, y_test_reg, color='blue', label='Actual')
plt.plot(X_test_reg, y_pred_reg, color='red', label='Predicted')
plt.title("Linear Regression - Actual vs Predicted")
plt.xlabel("Feature X")
plt.ylabel("Target y")
plt.legend()
plt.show()
```

## 2. Logistic Regression

    Logistic Regression is used for binary classification tasks. We'll train it on the classification dataset.

```python
# Split data into training and test sets

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_classification, y_classification, test_size=0.3, random_state=42)

# Initialize and train the Logistic Regression model
logistic_reg_model = LogisticRegression()

logistic_reg_model.fit(X_train_clf, y_train_clf)

# Make predictions
y_pred_clf_logistic = logistic_reg_model.predict(X_test_clf)

# Evaluate the model
accuracy = accuracy_score(y_test_clf, y_pred_clf_logistic)
print(f"Logistic Regression Accuracy: {accuracy}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_clf, y_pred_clf_logistic)
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d')

plt.title("Logistic Regression Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

plt.show()
```

# 3. Decision Tree Classifier

    Decision Trees are flexible models used for both classification and regression. Here, we'll apply it to our classification dataset.

```python
# Initialize and train the Decision Tree model
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(X_train_clf, y_train_clf)

# Make predictions
y_pred_clf_tree = decision_tree_model.predict(X_test_clf)

# Evaluate the model
accuracy_tree = accuracy_score(y_test_clf, y_pred_clf_tree)
print(f"Decision Tree Accuracy: {accuracy_tree}")

# Confusion Matrix
conf_matrix_tree = confusion_matrix(y_test_clf, y_pred_clf_tree)
sns.heatmap(conf_matrix_tree, annot=True, cmap='Greens', fmt='d')
plt.title("Decision Tree Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

# 4. Random Forest Classifier

    Random Forests are ensembles of decision trees that improve accuracy by combining the predictions of multiple trees.

```python
# Initialize and train the Random Forest model
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train_clf, y_train_clf)

# Make predictions
y_pred_clf_forest = random_forest_model.predict(X_test_clf)

# Evaluate the model
accuracy_forest = accuracy_score(y_test_clf, y_pred_clf_forest)
print(f"Random Forest Accuracy: {accuracy_forest}")

# Confusion Matrix
conf_matrix_forest = confusion_matrix(y_test_clf, y_pred_clf_forest)
sns.heatmap(conf_matrix_forest, annot=True, cmap='Purples', fmt='d')
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

## Step 4: Comparing the Models

    Print the accuracy of all the classification models to compare their performance.

```python
print("\nModel Comparison:")
print(f"Logistic Regression Accuracy: {accuracy}")
print(f"Decision Tree Accuracy: {accuracy_tree}")
print(f"Random Forest Accuracy: {accuracy_forest}")
```

## Unsupervised Learning Overview

    K-Means Clustering: A clustering algorithm that partitions data points into K distinct clusters based on feature similarity. Each data point is assigned to the cluster with the nearest centroid.

    Principal Component Analysis (PCA): A dimensionality reduction technique that transforms a dataset into a set of orthogonal components (principal components), capturing the maximum variance with fewer dimensions.

# Python Example: K-Means Clustering and PCA

Step 1: Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

Step 2: Create a Synthetic Dataset

We’ll create a synthetic dataset suitable for clustering and dimensionality reduction tasks.

```python
# Create a synthetic dataset with 3 clusters using make_blobs
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=42)

# Visualize the synthetic dataset
plt.scatter(X[:, 0], X[:, 1], c='gray', s=50)
plt.title("Synthetic Dataset for Clustering")
plt.xlabel("Feature X1")
plt.ylabel("Feature X2")
plt.show()
```

Step 3: K-Means Clustering

    1. Apply K-Means Clustering

    #Apply K-Means clustering with 3 clusters

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize the K-Means clustering results
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering Results")
plt.xlabel("Feature X1")
plt.ylabel("Feature X2")
plt.legend()
plt.show()
```

## Understanding the K-Means Output:

Data points are assigned to one of the K clusters, each represented by a unique color.

The red 'X' markers represent the centroids of each cluster, indicating the central point of each cluster.

2.Evaluating the Optimal Number of Clusters Using the Elbow Method

    The Elbow Method helps determine the optimal number of clusters by plotting the sum of squared distances from each point to its assigned cluster center (within-cluster sum of squares - WCSS).

```python
# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.show()
```

The "elbow" point in the graph indicates the optimal number of clusters (in this case, K = 3).

## Step 4: Principal Component Analysis (PCA)

PCA is useful for reducing the dimensionality of high-dimensional datasets while retaining as much variance as possible.

## 1. Apply PCA for Dimensionality Reduction

    Let's reduce a 3-dimensional synthetic dataset to 2 dimensions using PCA.

```python
# Create a 3-dimensional synthetic dataset

np.random.seed(42)

X_3d = np.dot(np.random.rand(3, 3), np.random.randn(3, 300)).T

# Visualize the original 3D data (only two dimensions plotted)
plt.scatter(X_3d[:, 0], X_3d[:, 1], c='gray', s=50)
plt.title("Original 3D Dataset (projected to 2D)")
plt.xlabel("Feature X1")
plt.ylabel("Feature X2")
plt.show()

# Apply PCA to reduce the data from 3D to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_3d)

# Visualize the reduced 2D dataset
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', s=50, alpha=0.7)
plt.title("PCA: Reduced 2D Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

## 2. Understanding the PCA Output:

    The transformed dataset now has only two dimensions (principal components) that capture most of the variance.

    PCA reduces the dataset's dimensionality while retaining as much information as possible.

# Model Evaluation and Validation in Machine Learning

    Model evaluation is crucial in determining a model's effectiveness and suitability for deployment, especially in embedded applications where performance, memory, and computation constraints are often critical

## 1. Model Evaluation Techniques

    Train/Test Split: A basic method where the dataset is divided into two sets – one for training and one for testing. The model is trained on the training set and evaluated on the test set.

    Cross-Validation: This is a more robust approach than the train/test split. Cross-validation helps ensure the model generalizes well to unseen data by performing multiple train/test splits.


## 2. Cross-Validation Techniques

    K-Fold Cross-Validation: The dataset is split into 'k' subsets (folds). The model is trained on k−1 folds and tested on the remaining fold. This process is repeated k times, with each fold serving as a test set once.

    Stratified K-Fold Cross-Validation: A variation of K-Fold where each fold maintains the same class distribution as the entire dataset. It’s commonly used for imbalanced datasets.

    Leave-One-Out Cross-Validation (LOOCV): Every sample in the dataset serves as a test set one by one, while the rest serve as the training set. It is computationally expensive but useful for small datasets.


## 3. Performance Metrics for Embedded Applications

In embedded systems, metrics are often constrained by memory and power requirements. Typical metrics include:

    Accuracy: Ratio of correct predictions to total predictions.

    Precision, Recall, and F1-Score: Particularly useful for classification problems, especially when classes are imbalanced.

    Inference Time: Measures how fast a model can make predictions, critical in real-time applications.

    Memory Usage: Indicates how much memory the model consumes. This is essential for devices with limited memory.

    Energy Consumption: Important for battery-operated or low-power devices.


## 4. Python Example: Model Evaluation with Cross-Validation and Metrics

    Here's an example using a Decision Tree Classifier with cross-validation and key metrics.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score, train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Load the Iris dataset
data = load_iris()
X, y = data.data, data.target

# Split the data into train and test sets for initial evaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree model
model = DecisionTreeClassifier()

# K-Fold Cross-Validation

k_fold = StratifiedKFold(n_splits=5)
cross_val_accuracy = cross_val_score(model, X, y, cv=k_fold, scoring='accuracy')

print(f'Cross-Validation Accuracy Scores: {cross_val_accuracy}')

print(f'Average Cross-Validation Accuracy: {np.mean(cross_val_accuracy)}')

# Train and evaluate the model on the test set

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

# 5. Explaining Metrics and Interpretations

    Cross-Validation Scores: Cross-validation helps confirm that the model performs consistently across different subsets of data. A high variance across folds may suggest overfitting.

    Accuracy: While useful, accuracy alone may not be adequate for imbalanced classes. It should be interpreted with other metrics.

    Precision: High precision indicates a low false positive rate, which is essential in embedded systems where erroneous classifications may lead to operational risks.

    Recall: High recall suggests a low false negative rate, beneficial in applications where missing a positive detection could be costly.

    F1 Score: A balanced metric combining precision and recall. A high F1 score means both false positives and false negatives are minimized.

    This approach ensures that the model is not only accurate but also meets performance requirements specific to embedded applications. By focusing on cross-validation and key metrics, the model evaluation process becomes more robust and reliable, particularly in resource-constrained environments.

# Model Evaluation and Validation Techniques for cross-validation in machine learning models 

    To thoroughly understand model evaluation and validation, 

    let’s walk through a step-by-step example of evaluating a machine learning model with cross-validation techniques and performance metrics, focusing on considerations for embedded applications.

    For this example, we will use a Decision Tree Classifier on the well-known Iris dataset.

## Step 1: Import Libraries and Load Data

    First, let’s import the necessary libraries and load the dataset.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Load the Iris dataset
data = load_iris()
X, y = data.data, data.target
```

## Step 2: Split Data for Initial Testing

    Splitting the data into training and testing sets allows us 
    to perform an initial evaluation before cross-validation. 
    This helps gauge the model’s performance on a holdout dataset.

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Step 3: Choose the Model

    For this example, we will use a Decision Tree Classifier.

```python
# Initialize the model
model = DecisionTreeClassifier()
```

## Step 4: Cross-Validation Technique – K-Fold Cross-Validation

### Cross-Validation with StratifiedKFold

    Cross-validation helps verify the model’s consistency across different subsets of data. 
    
    For this example, we’ll use Stratified K-Fold Cross-Validation, which maintains class distribution across folds, important for imbalanced data.


```python
# Stratified K-Fold Cross-Validation with 5 folds

k_fold = StratifiedKFold(n_splits=5)

cross_val_accuracy = cross_val_score(model, X, y, cv=k_fold, scoring='accuracy')

# Print cross-validation accuracy for each fold

print(f'Cross-Validation Accuracy Scores: {cross_val_accuracy}')
print(f'Average Cross-Validation Accuracy: {np.mean(cross_val_accuracy)}')
```

This provides an average score of the model’s performance across 5 different subsets of the data, giving us a more reliable estimate of its accuracy.
Step 5: Train and Test the Model on Holdout Set
After cross-validation, we can train the model on the entire training set and test it on the holdout test set. This step mimics real-world deployment, where unseen data is introduced to the model.


```python
# Train the model on the training set
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
```

## Step 6: Calculate Performance Metrics

    For embedded applications, evaluating with more than just accuracy is essential, as resources are often constrained. We’ll compute precision, recall, F1-score, and, optionally, inference time if evaluating embedded use.

    Key Metrics Explained
    1.Accuracy: The ratio of correct predictions to total predictions.
    2.Precision: Proportion of true positive predictions among all positive predictions.
    3.Recall: Proportion of true positives identified out of all actual positives.
    4.F1-Score: Harmonic mean of precision and recall, useful for imbalanced classes.

```python
# Metric Calculations in Python

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
# Calculate precision, recall, and F1 score (weighted for multi-class)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Display results
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

## Step 7: Interpret the Results

    Cross-Validation Scores: The cross_val_accuracy provides scores from each fold in cross-validation. If scores vary significantly across folds, the model might overfit on specific subsets of data.

    Accuracy: High accuracy suggests the model performs well on the given test set. However, accuracy alone may not indicate the model’s quality, especially if the dataset is imbalanced.

    Precision and Recall: High precision with moderate recall suggests the model has a low false positive rate but may miss some true positives, which may or may not be acceptable in embedded systems.

    F1-Score: A balanced metric combining precision and recall, useful if precision and recall hold equal importance for the application.


## Step 8: Evaluate Additional Embedded Application Constraints (Optional)

    For embedded applications, additional constraints may include:

    1.Inference Time: Measures how quickly the model predicts. This can be tested by timing the prediction loop and ensuring it meets the real-time requirements.

    2.Memory Usage: Indicates the model’s memory footprint. Libraries like memory_profiler or embedded-specific profiling tools can help.


    This step-by-step approach ensures robust model evaluation and validation, providing insights into accuracy and other key metrics essential for assessing the model’s suitability, especially when deploying to embedded systems.
